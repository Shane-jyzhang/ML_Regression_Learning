```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Import the Data We Need

```python
data = pd.read_csv("Used german cars.csv")

data
```

# Prepare the Data

1. Convert character variable encoding to numeric variable

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder() # initialize LabelEncoder

# Encode the entire DataFrame
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = encoder.fit_transform(data[column])
        
data
```

2. Split the X(features) and Y(target)

```python
x = data[['Make', 'Model', 'Build Year', 'Engine Size', 'Transmission', 'Mileage [km]', 'Fuel', 'Fuel efficiency']]
y = data[['Price [EUR]']]

x
```

3. Data set division

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train
```

4. Calculate the Pearson correlation index

相关系数的绝对值越接近于1，表示两个变量之间的线性相关性越强。如果相关系数接近于1（正或负），那么这两个变量之间的线性关系就越明显。当相关系数接近于1时，意味着一个变量的变化很可能会导致另一个变量的相应变化。因此，在某些情况下，你可以考虑保留其中一个变量，因为它们之间的信息可能有一定的重复性。但需要注意的是，仅仅通过相关系数来判断是否舍弃一列并不总是一个好的做法。有些情况下，即使相关系数很高，两个变量也可能承担着不同的角色或者在其他方面是重要的。

```python
correlation = np.corrcoef([data['Make'], data['Model'], data['Build Year'], data['Engine Size'], 
                           data['Transmission'], data['Mileage [km]'], data['Fuel'], data['Fuel efficiency']])
correlation = pd.DataFrame(correlation, 
                           columns=['Make', 'Model', 'Build Year', 'Engine Size', 'Transmission', 'Mileage [km]', 
                                    'Fuel', 'Fuel efficiency'],
                          index=['Make', 'Model', 'Build Year', 'Engine Size', 'Transmission', 'Mileage [km]', 
                                 'Fuel', 'Fuel efficiency'])

abs_values = correlation.abs()

abs_min_value = abs_values.min().min()

asb_second_largest_value = sorted(set(abs_values.values.flatten()), reverse=True)[3]

print(f"min_value = {abs_min_value}")
print(f"second_largest_value = {asb_second_largest_value}")
correlation
# 从输出来看，build year和mileage[km]相关性最强，为-0.765678。所以我们可以考虑将其中一列舍去。这个舍去的过程就是降维。
# Model和Transmission相关性最低，可以可视化看看。
```

5. Visualize the Data

```python
plt.figure(figsize=(10, 8))
plt.imshow(abs_values, cmap='YlGnBu', vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(range(len(abs_values.columns)), abs_values.columns, rotation=90)
plt.yticks(range(len(abs_values.index)), abs_values.index)
plt.title('Pearson Correlation Coefficients')
plt.show()
```

# Calculate MAPE

```python
def calculate_mape(y_true, y_pred):
    assert len(y_true) == len(y_pred), "Length of y_true and y_pred must be the same."
    n = len(y_true)
    mape = (1/n) * sum(abs((y_true[i] - y_pred[i]) / y_true[i]) for i in range(n)) * 100
    return mape
```

# Ridge Regression

1. Initialize and train the model

```python
from sklearn.linear_model import Ridge

model = Ridge(alpha=2.0)

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

y_true = y_test
y_true = y_true.values

print(y_pred)
print(y_true)
```

2. Evaluate the model

MAPE（Mean Absolute Percentage Error）是用于衡量回归模型在预测中的精确度的指标。它以百分比的形式表示误差程度，可以帮助你了解模型的预测准确度。MAPE 的值越小越好，它表示模型的平均相对误差百分比。

R平方（R-squared, R2）：R平方值度量了模型对目标变量的变异性解释程度，取值范围在0到1之间，越接近1表示模型拟合得越好。

均方误差（Mean Squared Error, MSE）：均方误差是预测值与实际值之间差的平方的平均值。MSE 值越小越好，表示模型对数据的拟合程度越好。同理均方根误差（Root Mean Squared Error, RMSE）

在机器学习中，准确率通常用于分类任务，而不是线性回归任务。线性回归任务的目标是预测一个连续的数值，而不是分类。

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"R2: {r2}")

mape = calculate_mape(y_true, y_pred)
print(f"MAPE: {mape}%")
```
