# Import and prepare the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Used german cars.csv")
# data
```

```python
# Encode the entire DataFrame
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder() # initialize LabelEncoder
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = encoder.fit_transform(data[column])
```

```python
x = data[['Make', 'Model', 'Build Year', 'Engine Size', 'Transmission', 'Mileage [km]', 'Fuel', 'Fuel efficiency']]
y = data[['Price [EUR]']]
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Neural Networks, NN

1.Use fully connected neural networks (also called multilayer perceptrons) to solve regression problems.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
# There are 6 layers in total. 1 input layer, 4 hidden layers and 1 output layer.
model = Sequential([
    Dense(128, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1)  # The output layer, because it is a regression problem, has only one output node.
])
# Compile model
model.compile(optimizer='adam', loss='mean_squared_error')
# Train model
model.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_test, y_test))
y_pred = model.predict(x_test)
```

2.Calculate the Mean Absolute Percentage Error (MAPE)

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape * 100:.2f}%")
```

3.Visualize the relationship between true and predicted values

```python
plt.scatter(range(len(y_test)), y_test, color='red', label='True Values')
plt.scatter(range(len(y_pred)), y_pred, color='blue', label='Predictions')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.legend()
plt.show()
```

# Decision Trees and Random Forests

1.A decision tree model and a random forest model were constructed and trained respectively.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# Initialize the decision tree model and train
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(x_train, y_train)
# Initialize the random forest model and train
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
# predict
y_pred_tree = tree_model.predict(x_test)
y_pred_rf = rf_model.predict(x_test)
```

2.Calculate the MAPE value

```python
mape_tree = mean_absolute_percentage_error(y_test, y_pred_tree)
mape_rf = mean_absolute_percentage_error(y_test, y_pred_rf)
print(f"MAPE for Decision Tree: {mape_tree * 100:.2f}%")
print(f"MAPE for Random Forest: {mape_rf * 100:.2f}%")
```

3.Visualize the true and predicted values of two models

```python
plt.scatter(y_test, y_test, color='red', label='True Values')
plt.scatter(y_test, y_pred_tree, color='green', label='Decision Tree Predictions')
plt.scatter(y_test, y_pred_rf, color='blue', label='Random Forest Predictions')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.legend()
plt.show()
```
